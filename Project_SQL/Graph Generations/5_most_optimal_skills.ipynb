```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from CSV results
data = {
    "skills": [
        "kafka", "pytorch", "perl", "tensorflow", "cassandra", "atlassian", "airflow", "scala", 
        "linux", "confluence", "pyspark", "mongodb", "mongodb", "gcp", "spark", "splunk", "databricks", 
        "git", "snowflake", "shell", "unix", "hadoop", "pandas", "phoenix", "php"
    ],
    "demand_count": [
        40, 20, 20, 24, 11, 15, 71, 59, 58, 62, 49, 26, 26, 78, 187, 15, 102, 74, 241, 44, 37, 140, 
        90, 23, 29
    ],
    "avg_salary": [
        129999.16, 125226.20, 124685.75, 120646.83, 118406.68, 117965.60, 116387.26, 115479.53, 
        114883.20, 114153.12, 114057.87, 113607.71, 113607.71, 113065.48, 113001.94, 112927.60, 
        112880.74, 112249.64, 111577.72, 111496.45, 111123.32, 110888.27, 110767.07, 109259.09, 109051.51
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Assign a unique color for each skill by setting a color palette
plt.figure(figsize=(12, 8))
sns.scatterplot(x="demand_count", y="avg_salary", data=df, hue="skills", palette="Set1", s=100)

# Add annotations for each point (skill names)
for i in range(df.shape[0]):
    plt.text(df["demand_count"].iloc[i] + 1, df["avg_salary"].iloc[i], df["skills"].iloc[i], 
             horizontalalignment='left', verticalalignment='bottom', fontsize=9)

# Title and labels
plt.title('Demand Count vs Average Salary by Skill', fontsize=16)
plt.xlabel('Demand Count', fontsize=12)
plt.ylabel('Average Salary ($)', fontsize=12)
plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()
```
