```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data from SQL query
data = {
    "job_title": [
        "Data Analyst", "Director of Analytics", "Associate Director- Data Insights", 
        "Data Analyst, Marketing", "Data Analyst (Hybrid/Remote)", "Principal Data Analyst (Remote)",
        "Director, Data Analyst - HYBRID", "Principal Data Analyst, AV Performance Analysis",
        "Principal Data Analyst", "ERM Data Analyst"
    ],
    "company_name": [
        "Mantys", "Meta", "AT&T", "Pinterest", "Uclahealthcareers", "SmartAsset", 
        "Inclusively", "Motional", "SmartAsset", "Get It Recruit"
    ],
    "salary_year_avg": [
        650000, 336500, 255829.5, 232423, 217000, 205000, 189309, 189000, 186000, 184000
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Sort by salary (descending)
df = df.sort_values(by="salary_year_avg", ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="salary_year_avg", y="job_title", hue="company_name", dodge=False, data=df, palette="Blues_r")

# Labels & Title
plt.xlabel("Salary (USD)")
plt.ylabel("")
plt.yticks(fontsize=7)
plt.title("Top 10 Data Analyst Salaries (Remote)")
plt.legend([],[], frameon=False)

# Show Plot
plt.show()

```
