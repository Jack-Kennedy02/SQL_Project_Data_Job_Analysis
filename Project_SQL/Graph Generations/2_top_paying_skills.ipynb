```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from SQL results
data = {
    "skills": [
        "sql", "python", "r", "azure", "databricks", "aws", "pandas", "pyspark", "jupyter", "excel", "tableau", "power bi", "powerpoint",
        "sql", "python", "r", "hadoop", "tableau", "sql", "crystal", "oracle", "tableau", "flow", "sql", "python", "go", "snowflake",
        "pandas", "numpy", "excel", "tableau", "gitlab", "sql", "python", "go", "snowflake", "pandas", "numpy", "excel", "tableau", 
        "gitlab", "sql", "python", "r"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Count of each skill
skill_counts = df['skills'].value_counts()

# Filter to include only skills with a count greater than 1
skill_counts = skill_counts[skill_counts > 1]

# Create a bar plot for the most common skills
plt.figure(figsize=(12, 8))
sns.barplot(x=skill_counts.values, y=skill_counts.index, palette="mako")
plt.title('Most Common Skills in the top 10 Highest Paying Data Analyst Roles')
plt.xlabel('Frequency')
plt.ylabel('Skills')
plt.show()
```
