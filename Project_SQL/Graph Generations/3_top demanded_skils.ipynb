```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from CSV results
data = {
    "skills": ["sql", "excel", "python", "tableau", "power bi"],
    "demand_count": [92628, 67031, 57326, 46554, 39468]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a bar plot for the skills demand
plt.figure(figsize=(10, 6))
sns.barplot(x="demand_count", y="skills", data=df, palette='mako')
plt.title('Skills Demand Count')
plt.xlabel('Demand Count')
plt.ylabel('')
plt.show()
```
